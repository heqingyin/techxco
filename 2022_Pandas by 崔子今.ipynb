### å…¶ä»–å¸¸ç”¨å‡½æ•°
Pandasè¿˜æä¾›äº†ä¸€äº›å¥½ç©ä¸”å®ç”¨çš„å‡½æ•°ï¼Œæ¯”å¦‚cumsum(), cumprod()

è¿™äº›å‡½æ•°ä¼šåˆ›å»ºä¸€ä¸ªæ–°çš„åˆ—ï¼Œä½ åº”è¯¥æŠŠè¿™ä¸ªåˆ—æ·»åŠ åˆ°åŸæœ¬çš„dataframeä¸­

```python
df['Age_cum'] = df['Age'].cumsum()
df.head()
```

cumsum(), cumulative sumçš„æ„ä¹‰æ˜¯å¯¹ä¸€ä¸ªåˆ—åšç´¯åŠ ï¼Œä¹Ÿå³Age_cum(t) = Agecum(t-1) + Age(t)ä¸æ–­çš„åšç´¯åŠ ï¼Œå¯ä»¥åœ¨é‡‘èé¢†åŸŸç®—ç´¯åŠ çš„ç®—æœ¯æ”¶ç›Š

cumprod()å‡½æ•°åˆ™ä¼šåšç´¯ä¹˜æ“ä½œï¼ŒAge_cum_prod(t) = Age_cum_prod(t) x Age(t)

```python
df['Age_cum_prod'] = df['Age'].cumprod()
df.head()
```

å…¶å®pandasè¿˜æ”¯æŒåˆ—ä¹‹é—´ç›´æ¥ä½¿ç”¨æ ‡å‡†è¿ç®—ç¬¦è¿›è¡Œæ“ä½œï¼Œæ¯”å¦‚+-*/

```python
df['temp'] = df['Age'] + df['Fare']
df.head()
```

```python
df['temp'] = df['Age'] / df['Fare']
df.head()
```

.apply()å‡½æ•°ä¼šæ¥å—ä¸€ä¸ªå‡½æ•°ï¼Œå¹¶ä¸”æŠŠè¿™ä¸ªå‡½æ•°ç”¨åœ¨DataFrameçš„æ‰€æœ‰è¡Œä¸­ï¼Œè¿™å°±æ˜¯loopçš„æ•ˆæœ

åœ¨è¿™é‡Œä½ å¯èƒ½ä¼šå‘ç°æˆ‘ä½¿ç”¨äº†lambdaï¼Œè¿™æ˜¯ä¸€ç§ç‰¹æ®Šè¡¨è¾¾å¼ï¼Œå¯ä»¥åˆ›é€ ä¸€ä¸ªç®€çŸ­çš„å‡½æ•°ï¼Œè¿™ä¸ªå‡½æ•°æ²¡æœ‰åå­—ï¼Œä¸å¯ä»¥å†è¢«å¼•ç”¨ï¼Œæ‰€ä»¥ä¸ä¼šæ±¡æŸ“å¤–éƒ¨ç©ºé—´ï¼ŒPythonä¸­çš„lambdaè¡¨è¾¾å¼å¯ä»¥ç›´æ¥ä¹¦å†™ç»“æœè€Œä¸å¿…ä½¿ç”¨returnå…³é”®å­—ã€‚

## 9. map
mapå‡½æ•°å¯ä»¥å¯¹dfä¸­çš„ä¸€æ•´åˆ—è¿›è¡Œæ˜ å°„ï¼Œ æ¯”å¦‚æˆ‘ä»¬ç°åœ¨å¸Œæœ›å°†dfçš„Sexé‚£ä¸€åˆ—ä¸­çš„maleæ ‡è®°ä¸º1ï¼Œfemaleæ ‡è®°ä¸º0ï¼Œè¿™åœ¨æ•°æ®é¢„å¤„ç†çš„æ—¶å€™æ˜¯éå¸¸å¸¸è§çš„æ“ä½œ
ç°åœ¨æˆ‘ä»¬æœ‰å¾ˆå¤šæ–¹æ³•å¯¹å…¶è¿›è¡Œå®ç°

```python
#é¦–å…ˆæ˜¯å®šä¹‰å‡½æ•°ä¹‹åçš„applyæ–¹æ³•
def convert_sex(x):
    if x == "male":
        return 1 
    else:   
        return 0

df["Sex"].apply(convert_sex)

#å…¶æ¬¡æ˜¯åŒ¿åå‡½æ•°çš„æ–¹æ³•
df["Sex"].apply(lambda x:  1 if x == "male" else 0)
```

```python
df["Sex"].map({"male": 1, "female":0})
```
