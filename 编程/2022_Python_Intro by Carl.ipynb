## å¾ªç¯ï¼š`for`
æˆ‘ä»¬åœ¨ä¸Šä¸€èŠ‚å­¦ä¹ äº†åˆ—è¡¨ `list`ã€‚å¦‚æœæˆ‘ä»¬è¦å¾ªç¯è¿‡ä¸€éè¿™ä¸ªåˆ—è¡¨ï¼ŒæŠŠæ‰€æœ‰å…ƒç´ æ‰“å°å‡ºæ¥ï¼Œæ€ä¹ˆåŠå‘¢ï¼Ÿ

```python
lights = ['green', 'yellow', 'red']

for light in lights:
    print(light)
```

åœ¨ä¸Šé¢çš„ä»£ç é‡Œé¢ï¼Œ`for light in lights` çš„æ„æ€ï¼Œæ˜¯ä¸€ä¸ªä¸€ä¸ªæŠŠ `lights` é‡Œé¢çš„å…ƒç´ æ”¾åˆ° `light` å˜é‡é‡Œæ¥æ‰§è¡Œä¸‹é¢çš„ä»£ç ã€‚å› ä¸ºåˆ—è¡¨çš„é•¿åº¦æ˜¯ 3ï¼Œæ‰€ä»¥ä¸‹é¢çš„ä»£ç  `print(light)` å°±æ‰§è¡Œäº† 3 æ¬¡ï¼Œæ¯ä¸€æ¬¡æ‰“å°ä¸€ä¸ªå…ƒç´ ã€‚

å¦‚æœæˆ‘ä»¬æƒ³è¦åœ¨å¾ªç¯çš„åŸºç¡€ä¸Šåšåˆ¤æ–­å‘¢ï¼Ÿæ¯”å¦‚è¯´ï¼Œå¯¹äºæ¯ä¸€ä¸ªä¿¡å·ç¯ `light`ï¼Œæˆ‘ä»¬éƒ½æƒ³çŸ¥é“è½¦é€Ÿæ˜¯å¤šå°‘ï¼Œæ€ä¹ˆåŠå‘¢ï¼Ÿé‚£æˆ‘ä»¬å¯ä»¥ç›´æ¥è¿›è¡Œä¿„ç½—æ–¯å¥—å¨ƒæ“ä½œï¼ŒæŠŠ `if` å¥—åœ¨ `for` å†…éƒ¨ï¼š

```python
for light in lights:
    if light == 'green':
        speed = 20
    else:
        speed = 0
    print(light, speed)
```

è¿™æ®µä»£ç é‡Œé¢ï¼Œ`for` è®©ä¸‹é¢çš„ä»£ç æ‰§è¡Œäº† 3 æ¬¡ï¼Œæ¯æ¬¡éƒ½è¿›è¡Œäº† `if` åˆ¤æ–­ï¼š
1. ç¬¬ä¸€æ¬¡ `light` æ˜¯ `'green'`ï¼Œæ‰€ä»¥è½¦é€Ÿ `20`
2. ç¬¬äºŒæ¬¡ `light` æ˜¯ `'yellow'`ï¼Œæ‰€ä»¥è½¦é€Ÿ `0`
3. ç¬¬ä¸‰æ¬¡ `light` æ˜¯ `'red'`ï¼Œæ‰€ä»¥è½¦é€Ÿ `0`

âš ï¸ æ³¨æ„ï¼šå¦‚æœè¦å¥—å¨ƒçš„è¯ï¼Œè®°å¾—è®© `if` å®Œå…¨åŒ…è£¹åœ¨ `for` é‡Œé¢ï¼Œä¹Ÿå°±æ˜¯è¯´åœ¨ `if`/`else` è¿™å‡ è¡Œçš„åŸºç¡€ä¸Šå†åŠ  4 ä¸ªç©ºæ ¼ã€‚

å¦‚æœæˆ‘æƒ³åœ¨ä¸€ä¸ª`if`è¯­å¥åˆ¤å®šå¤šç§ä¸åŒæƒ…å†µï¼Œè¯¥æ€ä¹ˆåŠå‘¢ï¼Ÿä½¿ç”¨`elif` (else if)ï¼

```python
for light in lights:
    if light == 'green':
        speed = 20
    elif light == 'yellow':
        speed = 5
    else:
        speed = 0
    print(light, speed)
```

è¿™æ—¶ï¼Œæˆ‘ä»¬å…ˆåˆ¤å®š`light`æ˜¯ä¸æ˜¯`green`ã€‚å¦‚æœæ˜¯`green`ï¼Œç»“æŸåˆ¤å®šã€‚å¦‚æœä¸æ˜¯ï¼Œå†å»åˆ¤å®šæ˜¯ä¸æ˜¯`yellow`ï¼Œä¸æ˜¯çš„è¯ï¼Œå†è¿›å…¥æœ€åä¸€ä¸ªæ¡ä»¶ï¼Œæ— è®ºæ˜¯`red`ï¼Œè¿˜æ˜¯å…¶ä»–çš„å­—ç¬¦ï¼Œéƒ½ä¼šç»™`speed`èµ‹äºˆ`0`çš„å€¼ã€‚

å¦‚æœæˆ‘ä»¬æƒ³åŒæ—¶è·å–è¿™ä¸ª`list`çš„å€¼å’Œä»–çš„`index`ï¼Œå¯ä»¥ä½¿ç”¨`enumerate(list)`: 

```python
for i, light in enumerate(lights): 
    print(i, light)
```

å¦‚æœæˆ‘ä»¬æƒ³è¦å¾ªç¯ä¸€ä¸ªå­—å…¸ï¼ŒåŒæ—¶è·å–ä»–çš„`key`å’Œ`value`ï¼Œåº”è¯¥æ€ä¹ˆå¼„å‘¢ï¼Ÿ

```python
for k, v in xacademy.items(): 
    print(k, v)
```

é‚£å¦‚æœæˆ‘ä»¬åªçŸ¥é“å¾ªç¯é•¿åº¦ `n`ï¼Œä¸çŸ¥é“å¾ªç¯çš„æ•°æ®ç»“æ„ï¼Œæ€ä¹ˆç”¨ `for` å¾ªç¯æ¥ `n` éå‘¢ï¼Ÿç­”æ¡ˆå°±æ˜¯ç”¨ `range(n)`ã€‚

```python
for i in range(3):
    print(i)
```

`range(n)` å°±ç­‰ä»·äºä¸€ä¸ªé•¿åº¦ä¸º `n` çš„åˆ—è¡¨ `[0, 1, ..., n-1]`ã€‚æ‰€ä»¥ `for i in range(n)` è‚¯å®šä¼šå¾ªç¯ `n` æ¬¡ã€‚

å½“ç„¶ï¼Œæˆ‘ä»¬ä¹Ÿå¯ä»¥ç”¨rangeå»å†™å‡ºè·Ÿåˆšæ‰å¾ªç¯`lights`çš„é‚£ä¸ªè¯­å¥ï¼ŒåŒæ ·æ•ˆæœï¼Œåªæ˜¯éº»çƒ¦ä¸€äº›ï¼š

```python
for i in range(len(lights)):
    if lights[i] == 'green':
        speed = 20
    elif lights[i] == 'yellow':
        speed = 5
    else:
        speed = 0
    print(lights[i], speed)
```

`range()` é‡Œé¢å¯ä»¥æ”¾å…¥ä¸æ­¢ä¸€ä¸ªå‚æ•°ã€‚
å¦‚æœåªæœ‰ä¸€ä¸ªå‚æ•°ï¼Œ`range(stop)`å°±ä¼šäº§ç”Ÿä¸€ä¸ªé•¿åº¦ä¸º `stop` çš„åˆ—è¡¨ `[0, 1, ..., n-1]`ã€‚

```python
for i in range(3, 100, 25):
    print(i)
```

è¿™æ ·æˆ‘ä»¬ä¹Ÿå¯ä»¥å€’è®¡æ—¶ï¼Œåªè¦æ˜¯`start`æ¯”`stop`è¦å¤§ï¼Œ`step`æ˜¯è´Ÿæ•°: 

```python
for i in range(10,0, -1): 
    print(i)
```

## å¾ªç¯ï¼š`while`

é™¤äº† `for` å¾ªç¯ï¼Œè¿˜æœ‰ä¸€ç§å¾ªç¯å« `while`ã€‚`while` çš„æ„æ€æ˜¯**å½“**ï¼Œä¹Ÿå°±æ˜¯å½“æŸä¸ªæ¡ä»¶æˆç«‹çš„æ—¶å€™å°±ä¸€ç›´å¾ªç¯æ‰§è¡Œã€‚

```python
speed = 0

while speed < 60:
    speed += 20
    
print(speed)
```

åœ¨ä¸Šé¢çš„ä»£ç ä¸­ï¼Œå½“ `speed < 60` çš„æ—¶å€™ï¼Œä¸‹é¢çš„ä»£ç  `speed += 20` å°±ä¸€ç›´è¢«å¾ªç¯æ‰§è¡Œã€‚äºæ˜¯é€Ÿåº¦å°±åŠ åˆ° 60ï¼Œç„¶åå°±ä¸ç»§ç»­åŠ é€Ÿäº†ã€‚æ˜¯ä¸æ˜¯éå¸¸ç›´è§‚å‘¢ï¼Ÿ

### ä¸€è¡Œ`for` å¾ªç¯

ç»å¸¸æˆ‘ä»¬å¯ä»¥ç”¨ä¸€è¡Œå°±èƒ½å†™å®Œçš„`for`å¾ªç¯æ¥å®šä¹‰æ–°çš„åˆ—è¡¨æˆ–è€…å­—å…¸ï¼Œä¹Ÿå¯ä»¥ç»“åˆ`if`è¯­å¥ä½¿ç”¨ï¼Œæ¯”å¦‚è¯´è¿™æ ·ï¼š

```python
print([i+10 for i in [1, 2, 3]])      # => [11, 12, 13]
print([x for x in [3, 4, 5, 6, 7] if x > 5])  # => [6, 7]

print({x for x in 'abcddeef' if x not in 'abc'})
print({x: x**2 for x in range(5)})
```

### `try`å’Œ`except`å¤„ç†å¼‚å¸¸

# ç±»`class`

æˆ‘ä»¬éœ€è¦ç”¨`class`å…³é”®è¯æ¥å®šä¹‰ä¸€ä¸ªç±»ã€‚ä¸€ä¸ªç±»æœ‰å¾ˆå¤šä¸ªå±æ€§ã€‚å®ƒè¢«è¿™ä¸ªç±»çš„æ‰€æœ‰å®ä¾‹æ‰€å…±äº«ï¼Œæ¯”å¦‚è¯´ä¸‹é¢çš„`species`ã€‚

```python
class Human:
    
    species = "H. sapiens"
    
    # åŸºæœ¬çš„åˆå§‹åŒ–å‡½æ•°ï¼Œå½“å«å‡ºè¿™ä¸ªç±»çš„ä¸€ä¸ªå®ä¾‹çš„æ—¶å€™ï¼Œè¿™ä¸ªä¼šè¢«è°ƒç”¨ã€‚
    def __init__(self, name):
        # å°†å‚æ•°åˆ†é…ç»™å®ä¾‹çš„åç§°å±æ€§
        self.name = name
        #  åˆå§‹åŒ–å±æ€§
        self._age = 0

    # å®ä¾‹æ–¹æ³•ï¼Œç¬¬ä¸€ä¸ªå‚æ•°æ€»æ˜¯selfï¼Œå°±æ˜¯è¿™ä¸ªå®ä¾‹å¯¹è±¡
    def say(self, msg):
        return f"{self.name}: {msg}"
    
    #æ³¨æ„ï¼Œå¦‚æœè¦è°ƒç”¨ è¿™ä¸ªç±»é‡Œé¢çš„å±æ€§çš„æ—¶å€™ï¼Œéœ€è¦å†™self.å±æ€§åç§°

    # ç±»æ–¹æ³•ï¼Œè¢«æ‰€æœ‰æ­¤ç±»çš„å®ä¾‹å…±ç”¨ã€‚ç¬¬ä¸€ä¸ªå‚æ•°æ˜¯è¿™ä¸ªç±»å¯¹è±¡ã€‚
    @classmethod
    def get_species(cls):
        return cls.species
    
    # é™æ€æ–¹æ³•ã€‚è°ƒç”¨æ—¶æ²¡æœ‰å®ä¾‹æˆ–ç±»çš„ç»‘å®šã€‚
    @staticmethod
    def grunt():
        return "*grunt*"
```

```python
# æ„é€ ä¸€ä¸ªå®ä¾‹
i = Human(name="Ian")
print(i.say("hi"))     # å°å‡º "Ian: hi"

j = Human("Joel")
print(j.say("hello"))  # å°å‡º "Joel: hello"

# è°ƒç”¨ä¸€ä¸ªç±»æ–¹æ³•
i.get_species()   # => "H. sapiens"

# æ”¹ä¸€ä¸ªå…±ç”¨çš„ç±»å±æ€§
Human.species = "H. neanderthalensis"
i.get_species()   # => "H. neanderthalensis"
j.get_species()   # => "H. neanderthalensis"

# è°ƒç”¨é™æ€æ–¹æ³•
Human.grunt()   # => "*grunt*"
```

```python
def bubble_sort(a): 
    for i in range(0, len(a)): 
        for j in range(0,len(a)-i-1): 
            if a[j]>a[j+1]:
                temp = a[j+1]
                a[j+1] = a[j] 
                a[j] = temp
                
    print(a)
```

```python
def selection_sort(a):
    for i in range(0,len(a)):
        k = i 
        temp = a[i] 
        for j in range(i,len(a)):
            if a[j] < temp:
                temp = a[j] 
                k = j
        a[k] = a[i] 
        a[i] = temp
    print(a)
```

```python
def partition(l, r, nums):
    pivot, ptr = nums[r], l
    for i in range(l, r):
        if nums[i] <= pivot:
            nums[i], nums[ptr] = nums[ptr], nums[i]
            ptr += 1
    nums[ptr], nums[r] = nums[r], nums[ptr]
    return ptr
 
def quicksort(l, r, nums):
    if len(nums) == 1:  
        return nums
    if l < r:
        pi = partition(l, r, nums)
        quicksort(l, pi-1, nums)  
        quicksort(pi+1, r, nums)  
    return nums
```

```python
graph = {
  'A' : ['B','C'],
  'B' : ['D', 'E'],
  'C' : ['F'],
  'D' : [],
  'E' : ['F'],
  'F' : []
}

visited = [] # List to keep track of visited nodes.
queue = []     #Initialize a queue

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)
    while queue:
        s = queue.pop(0) 
        print (s, end = " ") 
    for neighbour in graph[s]:
        if neighbour not in visited:
            visited.append(neighbour)
            queue.append(neighbour)

bfs(visited, graph, 'A')
```
