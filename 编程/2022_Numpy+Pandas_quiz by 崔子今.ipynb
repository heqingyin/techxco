    tips：你可能会使用到isnull()函数

4. Age在这份dataframe为numerical类型的变量，将numerical类型的变量转化为categorical类型的变量是一种常见的操作  
    请将二十岁及以下的乘客标记为young，六十岁及以上的乘客标记为old，其余乘客标记为middle，并统计三个年龄类型的乘客的数量


```python
# 1
titanic = pd.read_csv("2022_Python_basic/train.csv")
titanic.head(5)

# 2
titanic.sort_values(["Fare"], ascending=False)[0:10]


# 3
titanic[titanic["Age"].isnull()] = titanic['Age'].mean()

# 4
def age_convert(x):
    if(x <= 20):
        return "young"
    elif(x >= 60):
        return "old"
    else:
        return "middle"

titanic["Age"] =  titanic["Age"].apply(age_convert)
titanic.groupby("Age").count()

```
